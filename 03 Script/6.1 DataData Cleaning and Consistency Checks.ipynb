# 4. Data Cleaning 

## 4.1. Handling missing values

```python
# Check for missing values in the entire DataFrame
missing_data = df.isnull().sum()
print(missing_data)
```

<p><b>'bedrooms' column:</b><br>
Upon examining the dataset, it appears that most entries with missing values correspond to private rooms. Therefore, the analysis has decided to fill in the blanks with a value of 1.</p>

```python
df['bedrooms'].fillna(1, inplace=True)
```

<p><b>'bathrooms' column:</b><br>
# Given that there are only a few entries with missing values (14 out of 6636), 
# The analyst has opted to remove these rows, as the impact on the analysis is minimal.</p>


```python
df.dropna(subset=['bathrooms'], inplace=True)
```

<b>'reviews_scores_...' columns:</b><br>
Upon dataset examination, most missing values are primarily associated with listings with 0 availability and the analyst cannot reasonably impute these missing values. Hence, the analyst decides to exclude them to improve the quality of your analysis.</p>





```python
df.dropna(subset=['review_scores_accuracy'], inplace=True)
```

```python
# Remove duplicate rows:
df.drop_duplicates(inplace=True)
```

## 4.3. Data Transformation


<p><b>'prices' Column</b><br>

```python
# Remove the '$' sign and convert the 'price' column to float
df['price'] = df['price'].str.replace('[\$,]', '', regex=True).astype(float)
```

<p><b>'amenities' Column</b><br>
The 'amenities' column includes a textual list of provided amenities for each listing. To enhance the data's usability for analysis, the analyst has decided to split the text based on the delimiter (comma) and create new binary columns for each amenity. </p>

```python
# Parse amenities using a comma delimiter
# Remove curly braces and quotes
df['amenities'] = df['amenities'].str.replace('["]', '', regex=True)  
```

```python
# Split the modified 'amenities' column by commas ,. 
df['amenities'] = df['amenities'].str.split(',') 
```

```python
# Join the string using a pipe symbol ('|') 
# then converts this string into a DataFrame of binary columns.
df_amenities = df['amenities'].str.join('|').str.get_dummies()
```

```python
# view the df_amenities DataFrame
df_amenities.head()
```

```python
df_amenities.info()
```

```python
# Calculate the total number of amenities for each listing 
df['total_amenities'] = df_amenities.sum(axis=1)
```

```python
# Check the modified 'listings' DataFrame
df.info()
```
